```python
import matplotlib.pyplot as plt

# Specify the agent_id
agent_id = 32320

# Filter data for the specified agent
agent_data = call_df[call_df['Agent_ID'] == agent_id]

# Group by week and calculate average talk time for numeric columns only
weekly_avg_talk_time = agent_data.resample('W-Mon', on='Call_Date')['Talk_Time'].mean()

# Plotting the average weekly call time over time
plt.figure(figsize=(12, 6))
weekly_avg_talk_time.plot()
plt.title(f'Average Weekly Call Time for Agent {agent_id}')
plt.xlabel('Date')
plt.ylabel('Average Talk Time (seconds)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter data for August 2018 for the specified agent
august_data = agent_data[(agent_data['Call_Date'] >= '2018-08-01') & (agent_data['Call_Date'] <= '2018-08-31')]

# Plotting the daily call time for August 2018
plt.figure(figsize=(12, 6))
plt.bar(august_data['Call_Date'], august_data['Talk_Time'], color='lightcoral', edgecolor='black')
plt.title(f'Daily Call Time for Agent {agent_id} - August 2018')
plt.xlabel('Date')
plt.ylabel('Talk Time (seconds)')
plt.grid(axis='y')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter data for 2018-08-07 for the specified agent
specific_day_data = agent_data[agent_data['Call_Date'] == '2018-08-07']

# Plotting the call times for 2018-08-07
plt.figure(figsize=(12, 6))
plt.bar(specific_day_data.index, specific_day_data['Talk_Time'], color='lightgreen', edgecolor='black')
plt.title(f'Call Time for Agent {agent_id} on 2018-08-07')
plt.xlabel('Call Number')
plt.ylabel('Talk Time (seconds)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Setting the style for seaborn plots
sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(x=specific_day_data.index, y=specific_day_data['Talk_Time'], color='lightgreen')
plt.title(f'Call Time for Agent {agent_id} on 2018-08-07')
plt.xlabel('Call Number')
plt.ylabel('Talk Time (seconds)')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
# Setting the color palette based on unique emotions
palette = sns.color_palette("husl", len(specific_day_data['Emotion'].unique()))

plt.figure(figsize=(14, 7))
sns.barplot(x=specific_day_data.index, y=specific_day_data['Talk_Time'], hue=specific_day_data['Emotion'], palette=palette)
plt.title(f'Call Time and Emotion for Agent {agent_id} on 2018-08-07')
plt.xlabel('Call Number')
plt.ylabel('Talk Time (seconds)')
plt.legend(title='Emotion')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
from midiutil import MIDIFile
import numpy as np

# Normalize the weekly_avg_talk_time data to fit within a MIDI pitch range
min_pitch = 60  # Middle C
max_pitch = 80
normalized_data = np.interp(weekly_avg_talk_time, (weekly_avg_talk_time.min(), weekly_avg_talk_time.max()), (min_pitch, max_pitch)).astype(int)

# Create a new MIDI file with 1 track
midi = MIDIFile(1)
track = 0
time = 0
midi.addTrackName(track, time, "Weekly Talk Time")
midi.addTempo(track, time, 120)  # Set tempo to 120 beats per minute

# Add each data point as a note in the MIDI sequence
for pitch in normalized_data:
    midi.addNote(track, 0, pitch, time, 1, 100)  # channel=0, duration=1 (in beats), volume=100
    time += 1

# Save the MIDI file
midi_file_path = "/Users/sebastianbayne/Documents/WBS/Bootcamp/Final_Project /Datasets /weekly_talk_time_midi.mid"
with open(midi_file_path, "wb") as midi_file:
    midi.writeFile(midi_file)

midi_file_path
```

```python
from midiutil import MIDIFile
import numpy as np

# Normalize the Talk_Time data to fit within a MIDI pitch range
min_pitch = 60  # Middle C
max_pitch = 80
normalized_pitches = np.interp(specific_day_data['Talk_Time'], 
                               (specific_day_data['Talk_Time'].min(), specific_day_data['Talk_Time'].max()), 
                               (min_pitch, max_pitch)).astype(int)

# Map each unique emotion to a specific MIDI instrument
emotion_to_instrument = {
    "anger": 40,     # Violin
    "neutral": 0,    # Acoustic Grand Piano
    "happiness": 25, # Acoustic Guitar
    "sadness": 47,   # Orchestral Harp
    # ... Add other emotions and corresponding instruments as needed
}

# Ensure all unique emotions are covered in the dictionary
for emotion in specific_day_data['Emotion'].unique():
    if emotion not in emotion_to_instrument:
        print(f"Warning: Emotion '{emotion}' not mapped to an instrument. Please add it to the dictionary.")

# Create a new MIDI file
midi = MIDIFile(1)
track = 0
time = 0
midi.addTrackName(track, time, "Call Time and Emotion")
midi.addTempo(track, time, 120)

# Add each data point as a note in the MIDI sequence
for pitch, emotion in zip(normalized_pitches, specific_day_data['Emotion']):
    channel = emotion_to_instrument[emotion]
    midi.addNote(track, channel, pitch, time, 1, 100)
    time += 1

# Save the MIDI file
with open("/Users/sebastianbayne/Documents/WBS/Bootcamp/Final_Project /Datasets/call_time_emotion_midi.mid", "wb") as midi_file:
    midi.writeFile(midi_file)
```

```python
# Group by 'Call_Type_ID' and compute the mean for 'Talk_Time'
avg_talk_time = call_df.groupby('Call_Type_ID')['Talk_Time'].mean()

# Join with type_df to get the call type descriptions
result = avg_talk_time.reset_index().merge(type_df, on='Call_Type_ID')

# Display the result
print(result[['Call_Type_Description', 'Talk_Time']])
```

```python
# Group by 'Call_Type_ID' and compute the median for 'Talk_Time'
median_talk_time = call_df.groupby('Call_Type_ID')['Talk_Time'].median()

# Join with type_df to get the call type descriptions
result = median_talk_time.reset_index().merge(type_df, on='Call_Type_ID')

# Display the result
print(result[['Call_Type_Description', 'Talk_Time']])
```

```python
# Group by 'Call_Type_ID' and compute the standard deviation for 'Talk_Time'
std_dev_talk_time = call_df.groupby('Call_Type_ID')['Talk_Time'].std()

# Join with type_df to get the call type descriptions
result = std_dev_talk_time.reset_index().merge(type_df, on='Call_Type_ID')

# Display the result
print(result[['Call_Type_Description', 'Talk_Time']])
```
