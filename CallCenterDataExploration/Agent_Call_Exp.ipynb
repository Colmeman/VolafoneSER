```python
import matplotlib.pyplot as plt

# Specify the agent_id
agent_id = 32320

# Filter data for the specified agent
agent_data = call_df[call_df['Agent_ID'] == agent_id]

# Group by week and calculate average talk time for numeric columns only
weekly_avg_talk_time = agent_data.resample('W-Mon', on='Call_Date')['Talk_Time'].mean()

# Plotting the average weekly call time over time
plt.figure(figsize=(12, 6))
weekly_avg_talk_time.plot()
plt.title(f'Average Weekly Call Time for Agent {agent_id}')
plt.xlabel('Date')
plt.ylabel('Average Talk Time (seconds)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from midiutil import MIDIFile
import numpy as np

# Normalize the weekly_avg_talk_time data to fit within a MIDI pitch range
min_pitch = 60  # Middle C
max_pitch = 80
normalized_data = np.interp(weekly_avg_talk_time, (weekly_avg_talk_time.min(), weekly_avg_talk_time.max()), (min_pitch, max_pitch)).astype(int)

# Create a new MIDI file with 1 track
midi = MIDIFile(1)
track = 0
time = 0
midi.addTrackName(track, time, "Weekly Talk Time")
midi.addTempo(track, time, 120)  # Set tempo to 120 beats per minute

# Add each data point as a note in the MIDI sequence
for pitch in normalized_data:
    midi.addNote(track, 0, pitch, time, 1, 100)  # channel=0, duration=1 (in beats), volume=100
    time += 1

# Save the MIDI file
midi_file_path = "/Users/sebastianbayne/Documents/WBS/Bootcamp/Final_Project /Datasets /weekly_talk_time_midi.mid"
with open(midi_file_path, "wb") as midi_file:
    midi.writeFile(midi_file)

midi_file_path
```

```python
from midiutil import MIDIFile
import numpy as np

# Normalize the Talk_Time data to fit within a MIDI pitch range
min_pitch = 60  # Middle C
max_pitch = 80
normalized_pitches = np.interp(specific_day_data['Talk_Time'], 
                               (specific_day_data['Talk_Time'].min(), specific_day_data['Talk_Time'].max()), 
                               (min_pitch, max_pitch)).astype(int)

# Map each unique emotion to a specific MIDI instrument
emotion_to_instrument = {
    "anger": 40,     # Violin
    "neutral": 0,    # Acoustic Grand Piano
    "happiness": 25, # Acoustic Guitar
    "sadness": 47,   # Orchestral Harp
    # ... Add other emotions and corresponding instruments as needed
}

# Ensure all unique emotions are covered in the dictionary
for emotion in specific_day_data['Emotion'].unique():
    if emotion not in emotion_to_instrument:
        print(f"Warning: Emotion '{emotion}' not mapped to an instrument. Please add it to the dictionary.")

# Create a new MIDI file
midi = MIDIFile(1)
track = 0
time = 0
midi.addTrackName(track, time, "Call Time and Emotion")
midi.addTempo(track, time, 120)

# Add each data point as a note in the MIDI sequence
for pitch, emotion in zip(normalized_pitches, specific_day_data['Emotion']):
    channel = emotion_to_instrument[emotion]
    midi.addNote(track, channel, pitch, time, 1, 100)
    time += 1

# Save the MIDI file
with open("/Users/sebastianbayne/Documents/WBS/Bootcamp/Final_Project /Datasets/call_time_emotion_midi.mid", "wb") as midi_file:
    midi.writeFile(midi_file)
```
